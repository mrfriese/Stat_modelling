# Database

Put all your results in an SQLite3 database (remember, SQLite stores its databases as files in your local machine - make sure to create your database in your project's data/ directory!)

```python
import sqlite3
from sqlite3 import Error

def create_connection(path):
    connection = None
    try:
        connection = sqlite3.connect(path)
        print("Connection to SQLite DB successful")
    except Error as e:
        print(f"The error '{e}' occurred")

    return connection

connection = create_connection('Leipzig_bikes.sqlite')
```

```python
def execute_query(connection, query):
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        connection.commit()
        print("Query executed successfully")
    except Error as e:
        print(f"The error '{e}' occurred")
```

```python
create_bike_data_table = """
CREATE TABLE IF NOT EXISTS bike_data (
  id INTEGER PRIMARY KEY AUTOINCREMENT,
  name TEXT NOT NULL,
  rating FLOAT,
  Latitude FLOAT,
  Longitude FLOAT
);
"""
```

```python
execute_query(connection, create_bike_data_table)  
```

```python
results.to_sql('bike_data', connection, if_exists='replace')
```

```python
def execute_read_query(connection, query):
    cursor = connection.cursor()
    result = None
    try:
        cursor.execute(query)
        result = cursor.fetchall()
        return result
    except Error as e:
        print(f"The error '{e}' occurred")
```

```python
SQL_check = """ SELECT * FROM bike_data"""
```

```python
execute_read_query(connection, SQL_check)
```
